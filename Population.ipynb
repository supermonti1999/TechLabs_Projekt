```python
import numpy as np
import pandas as pd 
# Data extracted from  https://population.un.org/wpp/Download/Standard/CSV/ for the Total Population and the Population Density. 
df= pd.read_csv('WPP2019_TotalPopulationBySex.csv')
df.head(5)
#Filtering of the data for the relevant data
Filter_2020=df['Time']==2020 #& df['Variant']=='Medium')

Filter_2020_v1=(df['Variant']=='Medium') 

Data_2020=df[Filter_2020_v1& Filter_2020]

Data_2020.shape
Data_2020.head(50)

#Filterungen der Locations nach den L채ndern und Filterung der Regionen m체ssen aussortiert werden, wenn finale columns mit L채ndern 
#da ist, kann die Spalte selektiert werden und zur Filterung verwendet werden?, L채nder in ISO Codes umwandeln?
#Drop the unnecessary columns of the data
Data_2020.drop(['VarID','Variant','MidPeriod'], axis=1, inplace=True)

Data_2020.head()
#Explanation of Parameters 
#Population density per square kilometer(thousands)
```

```python
#Data set with Population divided into different age groups with a AgeGrpSpan of 5 Years. 

POP=pd.read_csv('Pop_Sex.csv')
POP.drop(['LocID','VarID',],axis=1, inplace=True)
#Furthermore unnessecary columns were droped and the Data is filtered with the required data
POP.shape
Filter_POP=POP['Time']==2020
Filter_POP_2=(POP['Variant']=='Medium') 

POP=POP[Filter_POP_2 & Filter_POP]


POP.head(50)

```

```python
#Insert data with the ISO Codes of the countrys in order to accomplish comparability between the data sets 
ISO = pd.read_csv('ISO3166.csv')
#Eliminate the unneccesary columns
ISO= ISO[['name','alpha-3']]
#rename columns
ISO.columns= ["country","Code_3_digits"]
#Merge ISOcodes with the dataframe in a new dataframe
POP_ISO= pd.merge(POP, ISO, left_on='Location', right_on='country') 
POP_ISO.drop(['country'], axis=1, inplace=True)

#Test for Nan Values, Not found
POP_ISO["Code_3_digits"].isnull().sum()

#Export the dataframe  as a csv because the initial data was too big to be uploaded to GitHub
POP_ISO.to_csv(r'C:\Users\Sebas\Google Drive\Techlabs Projekt\POP_ISO.csv',index=True)


```
