```python
#Data set with Population divided into different age groups with a AgeGrpSpan of 5 Years. 

POP=pd.read_csv('Pop_Sex.csv')
POP.drop(['LocID','VarID',],axis=1, inplace=True)
#Furthermore unnessecary columns were droped and the Data is filtered with the required data
POP.shape
Filter_POP=POP['Time']==2020
Filter_POP_2=(POP['Variant']=='Medium') 

POP=POP[Filter_POP_2 & Filter_POP]


POP.head(50)

```

```python
#Insert data with the ISO Codes of the countrys in order to accomplish comparability between the data sets 
ISO = pd.read_csv('ISO3166.csv')
#Eliminate the unneccesary columns
ISO= ISO[['name','alpha-3']]
#rename columns
ISO.columns= ["country","Code_3_digits"]
#Merge ISOcodes with the dataframe of the Population in a new dataframe
POP_ISO= pd.merge(POP, ISO, left_on='Location', right_on='country') 
POP_ISO.drop(['country'], axis=1, inplace=True)

#Test for Nan Values, Not found
POP_ISO["Code_3_digits"].isnull().sum()

#Export the dataframe  as a csv because the initial data was too big to be uploaded to GitHub
POP_ISO.to_csv(r'C:\Users\Sebas\Google Drive\Techlabs Projekt\POP_ISO.csv',index=True)

ISO.head()
```
